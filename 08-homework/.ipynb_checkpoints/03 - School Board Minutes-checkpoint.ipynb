````python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
````

````python
url = 'http://www.mineral.k12.nv.us/pages/School_Board_Minutes'
response = requests.get('http://www.mineral.k12.nv.us/pages/School_Board_Minutes')
doc = BeautifulSoup(response.text)
````

````python
print(doc.prettify())
````

````python
doc.find_all('a',href=True)
````

````python
files = doc.find_all('a',href=True)
rows = []
for file in files:
    if file.find('span'):
        row = {}
        row['url'] = url + file['href']
        row['date'] = file.find('span').text
        rows.append(row)
    
df = pd.DataFrame(rows)
df#['date'] = df['date'].dt.strftime("")
````

````python
# clean it up

# drop rows if 'date' is not a date
df = df[df.date.str.contains(', 2')]

# clean up 2108
df['date'] = df.date.str.replace('2108','2018')

# remove weird whitespace after date string
df['date'] = df.date.str.strip()
````

````python
# reformat dates
df['date'] = pd.to_datetime(df.date,format="%B %d, %Y")
df
````

````python
df.to_csv('minutes.csv')
````
